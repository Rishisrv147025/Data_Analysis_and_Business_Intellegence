# Prediction part using machine learning

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
label=LabelEncoder()
scale=StandardScaler()
```

```python
dataset['Student Name']=label.fit_transform(dataset['Student Name'])
dataset['Field of Study']=label.fit_transform(dataset['Field of Study'])
dataset['Specialization']=label.fit_transform(dataset['Specialization'])
```

```python
dataset['Unnamed: 0']=scale.fit_transform(dataset['Unnamed: 0'].values.reshape(-1,1))
dataset['Date of Birth']=scale.fit_transform(dataset['Date of Birth'].values.reshape(-1,1))
dataset['Year of Admission']=scale.fit_transform(dataset['Year of Admission'].values.reshape(-1,1))
dataset['Expected Year of Graduation']=scale.fit_transform(dataset['Expected Year of Graduation'].values.reshape(-1,1))
dataset['Current Semester']=scale.fit_transform(dataset['Current Semester'].values.reshape(-1,1))
dataset['Fees']=scale.fit_transform(dataset['Fees'].values.reshape(-1,1))
dataset['Discount on Fees']=scale.fit_transform(dataset['Discount on Fees'].values.reshape(-1,1))
```

```python
#Splitting the dataset to X and Y.
x=dataset[['Unnamed: 0','Student Name','Date of Birth','Field of Study','Year of Admission','Expected Year of Graduation','Current Semester','Specialization','Discount on Fees']]
y=dataset[['Fees']]
```

# **Splitting the dataset**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

# **Importing the machine learning model and fitting the dataset**

```python
#Importing the machine learning libraries
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from xgboost import XGBRegressor
```
